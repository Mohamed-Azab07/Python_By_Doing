```python
def write_csv(filename, data, fieldnames):
    with open(filename, "w", newline="", encoding="utf-8") as file:
        writer = csv.DictWriter(file, filednames=FIELDNAMES)
        writer.writeheader()
        writer.writerows(data)

def read_csv(filename):
    if not os.path.exists(filename):
        return []
    with open(filename, "r", encoding="utf-8") as file:
        reader = csv.DictReader(file)
        return [row for row in reader]

def backup_file(filename):
    backup = filename + ".bak"
    if os.path.exists(filename):
        shutil.copy2(filename, backup)
        print(f"‚úÖ Backed up {filename} to {backup}")
    return backup

def is_duplicate(data, new_emp):
    for emp in data:
        if emp["ID"] == str(new_emp["ID"]) or emp["Email"] == new_emp["Email"]:
            return True
    return False

def add_employee_safe(new_emp, filename, fieldnames):
    """Safely add new employee with validation and backup."""
    current = read_csv(filename)
    
    # Validate ID is int
    try:
        new_emp["ID"] = int(new_emp["ID"])
        new_emp["Salary"] = int(new_emp["Salary"])
    except ValueError:
        raise ValueError("ID and Salary must be integers.")
    
    # Convert to string for comparison (CSV stores everything as str)
    new_check = {k: str(v) for k, v in new_emp.items()}
    
    if is_duplicate(current, new_check):
        print("‚ùå Duplicate ID or Email. Employee not added.")
        return

    # Backup before overwrite
    backup_file(filename)
    
    # Add and rewrite
    current.append(new_check)
    write_csv(filename, current, fieldnames)
    print(f"‚úÖ Employee '{new_emp['Name']}' added successfully.")

def main():
    print("üîß Initializing employee database...\n")
    
    # Step 1: Write initial data
    write_csv(CSV_FILE, employees_data, FIELDNAMES)
    print(f"üìÅ Initial data written to {CSV_FILE}\n")

    # Step 2: Compute average salary
    data = read_csv(CSV_FILE)
    salaries = [int(emp["Salary"]) for emp in data]
    avg = sum(salaries) / len(salaries)
    print(f"üìä Average Salary: ${avg:,.2f}\n")

    # Step 3: Filter Engineering
    eng = [emp for emp in data if emp["Department"] == "Engineering"]
    print("üë®‚Äçüíª Engineering Team:")
    for e in eng:
        print(f"  - {e['Name']}")
    print()

    # Step 4: Sort by salary (desc)
    sorted_by_salary = sorted(data, key=lambda x: int(x["Salary"]), reverse=True)
    print("üèÜ Top 3 Highest Salaries:")
    for emp in sorted_by_salary[:3]:
        print(f"  {emp['Name']}: ${int(emp['Salary']):,}")
    print()

    # Step 5: Export HR team
    hr_team = [emp for emp in data if emp["Department"] == "HR"]
    write_csv(HR_FILE, hr_team, FIELDNAMES)
    print(f"üì§ HR team exported to {HR_FILE}\n")

    # Step 6: Add new employee (with safety)
    new_emp = {
        "ID": 18,
        "Name": "Marta",
        "Department": "Marketing",
        "Salary": 100000,
        "Email": "marta@example.com"
    }
    add_employee_safe(new_emp, CSV_FILE, FIELDNAMES)

    # Final count
    final_data = read_csv(CSV_FILE)
    print(f"\n‚úÖ Final employee count: {len(final_data)}")

main()
```

```python
import csv
import os
import shutil
from typing import List, Dict

# ================================
# 1. INITIAL DATA
# ================================
employees_data = [
    {"ID": 1, "Name": "Alice Johnson", "Department": "Engineering", "Salary": 95000, "Email": "alice@example.com"},
    {"ID": 2, "Name": "Bob Smith", "Department": "HR", "Salary": 62000, "Email": "bob@example.com"},
    {"ID": 3, "Name": "Carlos Ruiz", "Department": "Engineering", "Salary": 105000, "Email": "carlos@example.com"},
    {"ID": 4, "Name": "Dana Lee", "Department": "Marketing", "Salary": 72000, "Email": "dana@example.com"},
    {"ID": 5, "Name": "Elena Patel", "Department": "Finance", "Salary": 88000, "Email": "elena@example.com"},
    {"ID": 6, "Name": "Frank Miller", "Department": "Engineering", "Salary": 97000, "Email": "frank@example.com"},
    {"ID": 7, "Name": "Grace Kim", "Department": "HR", "Salary": 64000, "Email": "grace@example.com"},
    {"ID": 8, "Name": "Hugo Chen", "Department": "Finance", "Salary": 91000, "Email": "hugo@example.com"},
    {"ID": 9, "Name": "Iris Wong", "Department": "Marketing", "Salary": 70000, "Email": "iris@example.com"},
    {"ID": 10, "Name": "Jamal Thompson", "Department": "Engineering", "Salary": 110000, "Email": "jamal@example.com"},
    {"ID": 11, "Name": "Kira Adams", "Department": "Engineering", "Salary": 99000, "Email": "kira@example.com"},
    {"ID": 12, "Name": "Liam O‚ÄôConnor", "Department": "HR", "Salary": 61000, "Email": "liam@example.com"},
    {"ID": 13, "Name": "Mia Zhang", "Department": "Finance", "Salary": 87000, "Email": "mia@example.com"},
    {"ID": 14, "Name": "Noah Williams", "Department": "Marketing", "Salary": 73000, "Email": "noah@example.com"},
    {"ID": 15, "Name": "Olivia Davis", "Department": "Engineering", "Salary": 102000, "Email": "olivia@example.com"},
]

CSV_FILE = "employees.csv"
HR_FILE = "hr_team.csv"
FIELDNAMES = ["ID", "Name", "Department", "Salary", "Email"]


# ================================
# 2. HELPER FUNCTIONS
# ================================

def write_csv(filename: str, data: List[Dict], fieldnames: List[str]) -> None:
    """Write list of dicts to CSV with header."""
    with open(filename, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)

def read_csv(filename: str) -> List[Dict]:
    """Read CSV into list of dicts. Return empty list if file missing."""
    if not os.path.exists(filename):
        return []
    with open(filename, "r", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        return [row for row in reader]

def backup_file(filename: str) -> str:
    """Create a backup: employees.csv.bak"""
    backup = filename + ".bak"
    if os.path.exists(filename):
        shutil.copy2(filename, backup)
        print(f"‚úÖ Backed up {filename} to {backup}")
    return backup

def is_duplicate(data: List[Dict], new_emp: Dict) -> bool:
    """Check for duplicate ID or Email."""
    for emp in data:
        if emp["ID"] == str(new_emp["ID"]) or emp["Email"] == new_emp["Email"]:
            return True
    return False

def add_employee_safe(new_emp: Dict, filename: str, fieldnames: List[str]) -> None:
    """Safely add new employee with validation and backup."""
    current = read_csv(filename)
    
    # Validate ID is int
    try:
        new_emp["ID"] = int(new_emp["ID"])
        new_emp["Salary"] = int(new_emp["Salary"])
    except ValueError:
        raise ValueError("ID and Salary must be integers.")
    
    # Convert to string for comparison (CSV stores everything as str)
    new_check = {k: str(v) for k, v in new_emp.items()}
    
    if is_duplicate(current, new_check):
        print("‚ùå Duplicate ID or Email. Employee not added.")
        return

    # Backup before overwrite
    backup_file(filename)
    
    # Add and rewrite
    current.append(new_check)
    write_csv(filename, current, fieldnames)
    print(f"‚úÖ Employee '{new_emp['Name']}' added successfully.")


# ================================
# 3. MAIN WORKFLOW
# ================================

def main():
    print("üîß Initializing employee database...\n")
    
    # Step 1: Write initial data
    write_csv(CSV_FILE, employees_data, FIELDNAMES)
    print(f"üìÅ Initial data written to {CSV_FILE}\n")

    # Step 2: Compute average salary
    data = read_csv(CSV_FILE)
    salaries = [int(emp["Salary"]) for emp in data]
    avg = sum(salaries) / len(salaries)
    print(f"üìä Average Salary: ${avg:,.2f}\n")

    # Step 3: Filter Engineering
    eng = [emp for emp in data if emp["Department"] == "Engineering"]
    print("üë®‚Äçüíª Engineering Team:")
    for e in eng:
        print(f"  - {e['Name']}")
    print()

    # Step 4: Sort by salary (desc)
    sorted_by_salary = sorted(data, key=lambda x: int(x["Salary"]), reverse=True)
    print("üèÜ Top 3 Highest Salaries:")
    for emp in sorted_by_salary[:3]:
        print(f"  {emp['Name']}: ${int(emp['Salary']):,}")
    print()

    # Step 5: Export HR team
    hr_team = [emp for emp in data if emp["Department"] == "HR"]
    write_csv(HR_FILE, hr_team, FIELDNAMES)
    print(f"üì§ HR team exported to {HR_FILE}\n")

    # Step 6: Add new employee (with safety)
    new_emp = {
        "ID": 16,
        "Name": "Zoe Mitchell",
        "Department": "Marketing",
        "Salary": 75000,
        "Email": "zoe@example.com"
    }
    add_employee_safe(new_emp, CSV_FILE, FIELDNAMES)

    # Final count
    final_data = read_csv(CSV_FILE)
    print(f"\n‚úÖ Final employee count: {len(final_data)}")

if __name__ == "__main__":
    main()
```
