```python
import os
import pandas as pd
```

```python
directory = "excel_files"
filenames = os.listdir(directory)
print(filenames)
```

```python
x = '2. 100 Year 2024.xlsx'
print(x[-9:-5])
```

```python
filepaths_2024 = [os.path.join(directory, filename) for filename in filenames if filename[-9:-5] == '2024']
print(filepaths_2024)
```

```python
filepaths_2025 = [os.path.join(directory, filename) for filename in filenames if filename[-9:-5] == '2025']
print(filepaths_2025)
```

```python
files_2024 = [pd.read_excel(filepath) for filepath in filepaths_2024]
df_2024 = pd.concat(files_2024, ignore_index=True)
df_2024.to_excel("2024.xlsx", index=False)

files_2025 = [pd.read_excel(filepath) for filepath in filepaths_2025]
df_2025 = pd.concat(files_2025, ignore_index=True)
df_2025.to_excel("2025.xlsx", index=False)
```
