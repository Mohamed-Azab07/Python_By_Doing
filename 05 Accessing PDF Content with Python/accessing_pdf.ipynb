#### Word Occurances

```python
word_counter = {}

with pdfplumber.open("1. pdf1.pdf") as pdf:
    pages = pdf.pages
    for page in pages:
        text = page.extract_text()
        punc = ".,!{}()[]?"
        new_text = ""
        for char in text:
            if char not in punc:
                new_text += char
        words = new_text.split()
        for word in words:
            if word in word_counter:
                word_counter[word] += 1
            else:
                word_counter[word] = 1

print(word_counter)
```

```python
userinput = "the"
print(f"The word '{userinput}' occurs '{word_counter[userinput]}' times")
```
