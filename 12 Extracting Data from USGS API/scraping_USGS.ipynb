```python
import requests
import pandas as pd
```

```python
def get_data(startdate, enddate):
    url = "https://earthquake.usgs.gov/fdsnws/event/1/query"
    params = {
        "format": "geojson",
        "starttime": startdate,
        "endtime": enddate
    }
    response = requests.get(url, params=params)
    data = response.json()
    return data


raw_data = get_data("2024-07-30", "2024-07-31")
```

```python
def extract_data(raw_data):
    data = []
    features = raw_data["features"]
    for feature in features:
        data.append(feature["properties"]["place"])
    return data

print(extract_data(raw_data))
```
