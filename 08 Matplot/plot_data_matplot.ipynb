```python
import matplotlib.pyplot as plt
```

```python
file_path = "data.txt"
```

```python
with open(file_path, "r") as file:
    data = file.readlines()
    print(data)
data = [float(num.strip()) for num in data]
print(data)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data, marker="o", linestyle="-", color="blue")
plt.title("Data from Textfile")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import os
```

```python
directory = "files"
```

```python
filenames = os.listdir(directory)

data = []

for filename in filenames:
    filepath = os.path.join(directory, filename)
    with open(filepath, "r") as file:
        data.append(float(file.read()))

print(data)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data, marker="o")
plt.show()
```
