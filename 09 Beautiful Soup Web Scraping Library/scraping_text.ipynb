```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://forecast.weather.gov/MapClick.php?lat=40&lon=-74"
```

```python
response = requests.get(url)

content = response.text
print(content)
```

```python
soup = BeautifulSoup(content, "html.parser")
print(soup)
```

```python
temps = soup.find_all(class_="temp")
len(temps)
```

```python
periods = soup.find_all(class_="period-name")
len(periods)
```

```python
mydic = {}

i = 0
while i < len(temps):
    mydic[periods[i].get_text()] = temps[i].get_text()
    i += 2

print(mydic)
```
