```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("houses.csv")
df.head(10)
```

```python
fig = px.scatter(df, x="Size", y="Price", color="Location", title="House Price vs. House Size", hover_data={"Bedrooms": True, 
"Bathrooms": True, 
"Age": True, 
"Size": False
}, text="Location")

fig.update_traces(
    textposition='top center',  # or 'middle center', 'bottom right', etc.
    textfont_size=10
)

fig.show()
```
