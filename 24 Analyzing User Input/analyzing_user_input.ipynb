```python
userInput = "10 19 24 25 33 33".strip()
```

```python
print("Number Analysis Results: ")
print("-" * 25)

nums = userInput.split()

print(f"Total Numbers: {len(nums)}")
print(f"Sum of Numbers: {sum(int(num) for num in nums)}")
print(f"Range of Numbers: {max(int(num) for num in nums) - min(int(num) for num in nums)}")
print(f"Average Number: {sum(int(num) for num in nums) / len(nums)}")
```
