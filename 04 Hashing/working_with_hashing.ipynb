```python
import os
```

```python
directory = "files"
filepath = os.walk(directory)

def get_content(filepath):
    with open(filepath, "rb") as file:
        content = file.read()
        hash_value = hash(content)
        return hash_value


file_contents = {}

for root, dirs, files in filepath:
    for filename in files:
        filepath = os.path.join(root, filename)
        hashvalue = get_content(filepath)
        if hashvalue in file_contents:
            print("Duplicate", filepath)
            print("Original file", file_contents[hashvalue])
        else:
            file_contents[hashvalue] = filepath
```
