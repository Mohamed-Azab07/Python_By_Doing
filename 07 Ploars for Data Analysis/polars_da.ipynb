```python
import polars as pl
```

```python
df = pl.read_csv("1. cars.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
cly_d_df = df.filter(pl.col("cylinders") == 4)
cly_d_df
```

```python
df.shape
```

```python
cly_d_df.shape
```

```python
mean_horspower = df.group_by("cylinders").agg(pl.col("horsepower").max())
mean_horspower
```

```python
setosa_species = df.filter(pl.col("species") == "setosa")
setosa_species
```
