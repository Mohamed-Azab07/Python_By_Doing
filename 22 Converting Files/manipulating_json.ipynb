```python
import requests
```

```python
url = "https://raw.githubusercontent.com/arditsulceteaching/hosted_files/main/geo.json"
```

```python
response = requests.get(url)

data = response.json()
print(data)
```

```python
print(len(data))
```

```python
userInput = int(input(""))

quizzes = data["quizzes"]

for quiz in quizzes:
    for q in quiz["questions"]:
        if q["id"] == userInput:
            print(f"The Question is: {q["question"]}")
            for choice, ans in q["choices"].items():
                if ans == True:
                    print(f"The Answer of that Question is: {choice}")
```
