```python
import pandas as pd
```

```python
def read_excel_file(filepath):
    return pd.read_excel(filepath)

df = read_excel_file("europe.xlsx")
```

```python
def convert_to_csv(df):
    return df.to_csv("exported_csv.csv", index=False)

def convert_to_json(df):
    return df.to_json("exported_json.json", orient="records")
```

```python
convert_to_csv(df)
convert_to_json(df)
```
