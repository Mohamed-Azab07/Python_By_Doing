```python
import requests
```

```python
month = 10
day = 15
```

```python
url = f"https://history.muffinlabs.com/date/{month}/{day}"

response = requests.get(url)

```

```python
data = response.json()
print(data)
```

```python
print("What happened in that day !!! ")
print(f"Date: {data["date"]}")
print(f"Resources: {data["url"]}")
print(f"Event Year: {data["data"]["Events"][0]["year"]}")
print(f"Event Text: {data["data"]["Events"][0]["text"]}")
```
