```python
import requests
```

```python
cleaned_stations = []
with open("2. stations.txt", "r") as file:
    stations = file.readlines()
    for station in stations:
        cleaned_stations.append(station.strip())

print(cleaned_stations)
```

```python
def construct_filepath(filepath):
    filepath = f"{filepath}.csv.gz"
    return filepath

def construct_url(url, filepath):
    file_url = f"{url}{filepath}"
    return file_url

def download_file(file_url, filepath):
    response = requests.get(file_url)
    with open(filepath, "wb") as file:
        file.write(response.content)
```

```python
for st in cleaned_stations:
    filepath = construct_filepath(st)
    file_url = construct_url(url, filepath)
    download_file(file_url, filepath)
```
