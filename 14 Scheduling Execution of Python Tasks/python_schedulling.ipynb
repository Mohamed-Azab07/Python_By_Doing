```python
import schedule
import datetime
import csv
import time
import os
```

```python
CSV_FILE = "timestamps.csv"

def append_current_time():
    now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    file_exists = os.path.isfile(CSV_FILE)
    with open(CSV_FILE, mode='a', newline='') as file:
        writer = csv.writer(file)
    
        if not file_exists:
            writer.writerow(["Timestamp"])
        writer.writerow([now])
    print(f"Appended: {now}")

schedule.every().minute.do(append_current_time)
# Run initial log (optional)
append_current_time()

# Keep the script running
print("Scheduler started. Logging time every minute...")
while True:
    schedule.run_pending()
    time.sleep(1)
```
