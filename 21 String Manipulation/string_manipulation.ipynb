#### Text Manipulation

```python
with open("snowwhite_test.txt", "r") as file:
    content = file.read()
    sentences = content.split(". ")
    cap_first_letter = [s.capitalize() for s in sentences]
    new_content = ". ".join(cap_first_letter)
    print(new_content)

with open("snowwhite_mod.txt", "w") as file:
    file.write(new_content + ".")
```
